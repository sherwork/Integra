```python
# расчет записей в колонке и уникальных записей в колонке
print ('название колонки - всего записей в колонке / уникальных записей в колонке:\n')
for i, col in enumerate(data.columns):
    print (i,')', col,'-', len(data[col]), '/', len(set(data[col])))
```

```python
# расчет количества каждой позиции в колонке
col = data['Статус']
for mark in np.unique(col.values):
    print (mark, ':', ' '*5, (col == mark).sum())
```

```python
col = data[u'Товар']
for mark in np.unique(col.values):
    print (mark, ':', ' '*5, (col == mark).sum())
```

```python
col = data['Система оплаты']
for mark in np.unique(col.values):
    print (mark, ':', ' '*5, (col == mark).sum())
```
